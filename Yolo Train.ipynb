```python
from ultralytics import YOLO
```

```python
# Load a pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')  # 'yolov8n.pt' (nano), 'yolov8s.pt' (small), 'yolov8m.pt' (medium)

# Train the model

model.train(
    data='dataset.yaml',
    epochs=50,
    imgsz=416,
    batch=16,
    name='sjhacks_yolo_model_fast',
    project='YOLOv8_training',
    pretrained=True,
    patience=5   # if no improvement after 5 epochs, stop early
)
```

  Best model found at epoch 14!

```python
# Load the best model
model = YOLO('/content/drive/MyDrive/SJHacks Hackathon/YOLOv8_training/sjhacks_yolo_model/weights/best.pt')

# Run validation
metrics = model.val()

# Run testing on test set
results = model.predict(
    source='/content/drive/MyDrive/SJHacks Hackathon/SplitDataset/Test/Images',
    save=True
)
```
